```python
numeric_ids = pd.to_numeric(df_movies['id'], errors='coerce')
invalid_ids_count = numeric_ids.isnull().sum()
print(f"número de IDs não numéricos encontrados: {invalid_ids_count}")

duplicate_ids_count = df_movies['id'].duplicated().sum()
print(f"número de IDs duplicados encontrados: {duplicate_ids_count}")
```

```python
# Transformação 1.0

# Limpa os IDs inválidos
df_movies['id'] = pd.to_numeric(df_movies['id'], errors='coerce')
df_movies.dropna(subset=['id'], inplace=True)
df_movies['id'] = df_movies['id'].astype(int)
df_credits['id'] = df_credits['id'].astype(int)
df_keywords['id'] = df_keywords['id'].astype(int)


# Unificandos os DataFrames
df_merged = pd.merge(df_movies, df_credits, on='id', how='left')
df_merged = pd.merge(df_merged, df_keywords, on='id', how='left')

# Removendo IDs duplicados
df_merged.drop_duplicates(subset=['id'], keep='first', inplace=True)


print("\nDataFrames mesclados com sucesso")
print(f"Formato final {df_merged.shape}")

print("\nInformações do DataFrame Mesclado")
df_merged.info()

print("\n5 Primeiras Linhas do DataFrame Mesclado")
display(df_merged.head())
```

```python
# Transformação 2.0 (conversão dos tipos de dados)

df_merged['budget'] = pd.to_numeric(df_merged['budget'], errors='coerce').fillna(0).astype(int)
print("Coluna 'budget' convertida para inteiro.")

df_merged['popularity'] = pd.to_numeric(df_merged['popularity'], errors='coerce').fillna(0).astype(float)
print("Coluna 'popularity' convertida para float.")

df_merged['release_date'] = pd.to_datetime(df_merged['release_date'], errors='coerce')
print("Coluna 'release_date' convertida para data.")

df_merged['adult'] = df_merged['adult'] == 'True'
print("Coluna 'adult' convertida para booleano.")

df_merged['video'] = df_merged['video'] == 'True'
print("Coluna 'video' convertida para booleano.")

print("\nTipos de dados após a conversão")
print(df_merged[['budget', 'popularity', 'release_date']].dtypes)
```

```python
# Função Auxiliar de Extração

import ast

def extract_json_data(data, key_to_extract, limit=None):
    
    if isinstance(data, str) and data.startswith('['):
        try:
            list_of_items = ast.literal_eval(data)

            if list_of_items:
                if limit:
                    list_of_items = list_of_items[:limit]

                names = [item.get(key_to_extract, '') for item in list_of_items]

                return ', '.join(filter(None, names))
        except (ValueError, SyntaxError):
            return ''
    return ''
```

```python

# Extração para Limpar as Colunas em JSON

df_merged['genres'] = df_merged['genres'].apply(lambda x: extract_json_data(x, 'name'))

df_merged['cast'] = df_merged['cast'].apply(lambda x: extract_json_data(x, 'name', limit=3))

df_merged['keywords'] = df_merged['keywords'].apply(lambda x: extract_json_data(x, 'name'))

def get_director(crew_data):
    if isinstance(crew_data, str) and crew_data.startswith('['):
        try:
            crew_list = ast.literal_eval(crew_data)
            for member in crew_list:
                if member.get('job') == 'Director':
                    return member.get('name', '')
        except (ValueError, SyntaxError):
            return ''
    return ''

df_merged['director'] = df_merged['crew'].apply(get_director)

print("extração concluída")

print("\ndados transformados")
display(df_merged[['title', 'genres', 'cast', 'director', 'keywords']].head())
```

```python
# Transformação final e garantindo alinhamento ao schema

def get_collection_name(data):
    if isinstance(data, str) and data.startswith('{'):
        try:
            collection_dict = ast.literal_eval(data)
            return collection_dict.get('name', '')
        except (ValueError, SyntaxError):
            return ''
    return ''

df_merged['belongs_to_collection'] = df_merged['belongs_to_collection'].apply(get_collection_name)
df_merged['production_companies'] = df_merged['production_companies'].apply(lambda x: extract_json_data(x, 'name', limit=3))
df_merged['production_countries'] = df_merged['production_countries'].apply(lambda x: extract_json_data(x, 'name'))
df_merged['spoken_languages'] = df_merged['spoken_languages'].apply(lambda x: extract_json_data(x, 'name'))

colunas_finais_movies = [
    'id', 'title', 'overview', 'release_date', 'budget', 'revenue', 'runtime',
    'popularity', 'status', 'tagline', 'vote_average', 'vote_count', 'imdb_id',
    'original_language', 'genres', 'production_companies', 'production_countries',
    'spoken_languages', 'belongs_to_collection',
    'director',
    'cast',
    'keywords'
]

df_movies_final = df_merged[colunas_finais_movies].copy()

for col in df_movies_final.select_dtypes(include='object').columns:
    df_movies_final[col] = df_movies_final[col].fillna('')

print("DataFrame final pronto")
df_movies_final.info()
```

```python
# Limpando tabela para inserção

from sqlalchemy import text

try:
    with engine.connect() as connection:
        with connection.begin() as transaction:
            connection.execute(text("SET FOREIGN_KEY_CHECKS = 0;"))
            
            print("Limpando a tabela 'ratings'...")
            connection.execute(text("TRUNCATE TABLE ratings;"))
            
            print("Limpando a tabela 'movies'...")
            connection.execute(text("TRUNCATE TABLE movies;"))
            
            connection.execute(text("SET FOREIGN_KEY_CHECKS = 1;"))
            
            print("Tudo limpo")            
except Exception as e:
    print(f"erro ao limpar as tabelas: {e}")
```

```python
# Carregando os dados na tabela movies
try:
    df_movies_final.to_sql(
        'movies',
        con=engine,
        if_exists='append', 
        index=False,     
        chunksize=1000    
    )
    print("Carregamento da tabela concluído")

except Exception as e:
    print(f"Ocorreu um erro durante o carregamento: {e}")
```

```python
# PipeLine de ETL para tabela ratings

try:
    df_ratings = pd.read_csv('ratings_small.csv')
    
    df_ratings.rename(columns={
        'userId': 'user_id',
        'movieId': 'movie_id',
        'timestamp': 'rating_timestamp'
    }, inplace=True)
    
    df_ratings['rating_timestamp'] = pd.to_datetime(df_ratings['rating_timestamp'], unit='s')
    
    valid_movie_ids = df_movies_final['id'].unique()
    
    df_ratings_valid = df_ratings[df_ratings['movie_id'].isin(valid_movie_ids)]
    
    print(f"Encontradas {len(df_ratings)} avaliações no CSV. {len(df_ratings_valid)} são.")
    
    df_ratings_valid.to_sql(
        'ratings',
        con=engine,
        if_exists='append',
        index=False,
        chunksize=5000
    )
    print("carregamento da tabela")

except Exception as e:
    print(f"ocorreu um erro durante o carregamento da tabela: {e}")
```

```python
# 2. Distribuição de filmes por década
df_movies_db['year'] = pd.to_datetime(df_movies_db['release_date']).dt.year
df_movies_db['decade'] = (df_movies_db['year'] // 10) * 10

print("\n" + "=" * 80)
print("DISTRIBUIÇÃO DE FILMES POR DÉCADA")
print("=" * 80)

decade_counts = df_movies_db['decade'].value_counts().sort_index()
print(decade_counts)

print(f"\n📅 Período coberto: {df_movies_db['year'].min():.0f} até {df_movies_db['year'].max():.0f}")
```

```python
# 5. Análise de gêneros mais comuns
print("\n" + "=" * 80)
print("ANÁLISE DE GÊNEROS")
print("=" * 80)

# Extrair todos os gêneros (estão separados por vírgula)
all_genres = []
for genres_str in df_movies_db['genres'].dropna():
    if genres_str:
        genres_list = [g.strip() for g in genres_str.split(',')]
        all_genres.extend(genres_list)

genre_counts = pd.Series(all_genres).value_counts().head(15)
print("\nTop 15 gêneros mais frequentes:")
print(genre_counts)
```

```python
# 7. Filmes mais avaliados
print("\n" + "=" * 80)
print("TOP 10 FILMES MAIS AVALIADOS")
print("=" * 80)

ratings_count = df_ratings_db.groupby('movie_id').agg({
    'rating': ['count', 'mean']
}).reset_index()
ratings_count.columns = ['movie_id', 'num_ratings', 'avg_rating']

# Join com a tabela de filmes para pegar os títulos
top_rated = ratings_count.nlargest(10, 'num_ratings')
top_rated = top_rated.merge(df_movies_db[['id', 'title']], left_on='movie_id', right_on='id')
display(top_rated[['title', 'num_ratings', 'avg_rating']])
```
